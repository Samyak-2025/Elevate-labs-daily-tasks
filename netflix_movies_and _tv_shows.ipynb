```python
## checking cleaning 
import pandas as pd

# Step 1: Load the data
df_raw = pd.read_csv("netflix_titles_nov_2019.csv")

# Step 2: Make a copy so we have the original for comparison
df_cleaned = df_raw.copy()

# Step 3: Count rows before cleaning
before_duplicates = df_cleaned.shape[0]

# Step 4: Remove duplicates
df_cleaned = df_cleaned.drop_duplicates()
after_duplicates = df_cleaned.shape[0]
duplicates_removed = before_duplicates - after_duplicates

# Step 5: Count rows before removing nulls
before_nulls = df_cleaned.shape[0]

# Step 6: Fill and drop nulls
df_cleaned['director'] = df_cleaned['director'].fillna('Not Available')
df_cleaned['cast'] = df_cleaned['cast'].fillna('Not Available')
df_cleaned['country'] = df_cleaned['country'].fillna('Unknown')
df_cleaned = df_cleaned.dropna(subset=['date_added', 'rating', 'duration'])

after_nulls = df_cleaned.shape[0]
nulls_removed = before_nulls - after_nulls

# Step 7: Print the summary
print(f"üîÅ Duplicate rows removed: {duplicates_removed}")
print(f"üï≥Ô∏è Rows removed due to nulls in key columns: {nulls_removed}")
print(f"üìâ Final dataset shape: {df_cleaned.shape}")
```
